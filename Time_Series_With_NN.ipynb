```python
import pandas as pd
```

```python
df = pd.read_csv("train_df.csv", index_col = 'id')
test =pd.read_csv('test_df.csv', index_col = 'id')
```

```python
df.head()
```

```python
import keras
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense, Dropout, Conv1D, MaxPooling1D
from keras.engine.input_layer import Input
from keras.callbacks import TerminateOnNaN
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop('y', axis = 1), df.y)
```

```python
nn = Sequential()
nn.add(Dense(128,  input_dim=500))
nn.add(Dropout(0.5))
nn.add(Dense(64, activation='relu'))
nn.add(Dropout(0.5))
# nn.add(Input(shape = 500))
# nn.add(Conv1D(32, 3, padding="same", activation="relu", input_shape = X_train.shape))
# nn.add(MaxPooling1D(pool_size=3))
nn.add(Dense(1, activation = 'sigmoid'))
# nn.add(Dense(2, activation='softmax'))
nn.compile(loss='binary_crossentropy', optimizer='Nadam', metrics=['accuracy'])
```

```python
history = nn.fit(X_train, y_train, batch_size=32, epochs=40, 
                      verbose=2, validation_data=[X_test, y_test])
```

```python
pred = nn.predict_proba(test)
```

```python
pred2=[]
for i in pred:
    pred2.append(i[0])
```
